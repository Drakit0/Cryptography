```python
def arreglartexto(texto):
    """
    Esta función elimina los espacios, las comas, puntos y comas, etc. También sustituye las vocales
    acentuadas por vocales sin acento y las eñes por enes.
    """
    texto = texto.lower()
    nuevotexto = ''
    for caracter in texto:
        if caracter in 'bcdfghjklmpqrstvwxyz':
            nuevotexto += caracter
        elif caracter in 'aáàâä':
            nuevotexto += 'a'
        elif caracter in 'eèéêë':
            nuevotexto += 'e'
        elif caracter in 'iìíîï':
            nuevotexto += 'i'
        elif caracter in 'oòóôö':
            nuevotexto += 'o'
        elif caracter in 'uùúûü':
            nuevotexto += 'u'
        elif caracter in 'nñ':
            nuevotexto += 'n'
    return  nuevotexto

def contar_caracteres(texto):
    """
    Devuelve un diccionario con las frecuencias de los distintos caracteres.
    """
    return Counter(texto)

    """"
    Este apartado lo iba a usar en cifrado monoalfabético, pero al final no fue necesario.
    """

    def sumar_caracteres(c1,c2):
        """
        Suma los caracteres módulo 26.
        """
        alfabeto = 'abcdefghijklmnopqrstuvwxyz'
        return alfabeto[(alfabeto.index(c1)+alfabeto.index(c2))%26]

    def restar_caracteres(c1,c2):
        """
        Resta los caracteres módulo 26.
        """
        alfabeto = 'abcdefghijklmnopqrstuvwxyz'
        return alfabeto[(alfabeto.index(c1)-alfabeto.index(c2))%26]    

def indice_coincidencia(texto):
    """
    Dado un texto devuelve el índice de coincidencia, es decir la probabilidad de que 
    dos caracteres elegidos al azar sean iguales.
    """
    texto = arreglartexto(texto)
    longitud = len(texto)
    counter = contar_caracteres(texto)
    indice = 0
    for key,value in counter.items():
        indice += (value/longitud)**2
    return indice

def frecuencias(texto,nombre='00'):
    """
    Dibuja un diagrama de barras con las frecuencias del texto.
    Exporta el diagrama de barras como imagen.
    """
    texto = arreglartexto(texto)
    dic = contar_caracteres(texto)
    orddic = sorted(dic.items())
    x,y = [],[]
    for i,j in orddic:
        x.append(i)
        y.append(j/len(texto))
    plt.xlabel('caracteres')
    plt.ylabel('frecuencias')
    plt.title(nombre)
    plt.grid()
    plt.bar(x,y,width=0.5)
    plt.savefig('frecuencias_'+nombre+'.png',dpi=180)
    plt.show()

def escribir_texto5(texto):
    """
    Esto es para escribir el texto cifrado. Lo escribe en mayúsculoa y en bloques de 5
    caracteres separados por un espacio.
    """
    texto = texto.upper()
    nuevo_texto = ''
    limite = len(texto)-len(texto)%5
    for k in range(0,limite,5):
        nuevo_texto += texto[k:k+5]
        nuevo_texto += ' '
    nuevo_texto += texto[limite:]
    return nuevo_texto

def cifradoCesar(texto,desplazamiento):
    """
    Devuelve el texto cifrado de Cesar.
    """
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'
    texto = arreglartexto(texto)
    texto_cifrado = ''
    for caracter in texto:
        texto_cifrado += alfabeto[(alfabeto.index(caracter)+desplazamiento)%26]
    return texto_cifrado

def cifrado_monoalfabetico(texto,alfabeto):
    """
    Devuelve el texto cifrado con el alfabeto dado.
    """
    alf = 'abcdefghijklmnopqrstuvwxyz'
    texto = arreglartexto(texto)
    texto_cifrado = ''
    for caracter in texto:
        texto_cifrado += alfabeto[alf.index(caracter)]
    return texto_cifrado

def cifradoVigenere(texto,clave):
    """
    Devuelve el cifrado de Vigenère con la clave dada.
    """
    texto = arreglartexto(texto)
    clave = arreglartexto(clave)
    texto_cifrado = ''
    contador = 0
    longitud = len(clave)
    for caracter in texto:
        texto_cifrado += sumar_caracteres(caracter,clave[contador%longitud])
        contador += 1
    return texto_cifrado

def trigrafos(texto):
    """
    Devuelve un diccionario con las frecuencias de los trígrafos en el texto.
    """
    texto = arreglartexto(texto)
    dicc = {}
    for k in range(len(texto)-2):
        trigrafo = texto[k:k+3]
        if trigrafo not in dicc:
            dicc[trigrafo] = 1
        else:
            dicc[trigrafo] += 1
    return {k:v for k,v in sorted(dicc.items(), key=lambda item:item[1],reverse=True)}


def trigrafos_frecuentes(texto,numero):
    """
    Devuelve un diccionario con las frecuencias de los trígrafos 
    más frecuentes. La cantidad de trígrafos que queremos ver está dado por la variable número .
    """
    dicc = trigrafos(texto)
    valor = 1
    while True:
        a = [(k,v) for k,v in dicc.items() if v>valor]
        if len(a)>numero:
            valor += 1
        else:
            return a


def longitudClaveVigenere(texto_cifrado):
    """
    Dado un texto cifrado por el método de Vigenère devuelve la longitud de la clave gracias a los trígrafos frecuentes.
    """
    ic = 0
    longitud = 0
    while ic<0.06:
        longitud += 1
        texto = ''
        for k in range(len(texto_cifrado)):
            if k%longitud==0:
                texto += texto_cifrado[k]
        ic = indice_coincidencia(texto)
    return longitud
        
def indice_longitud(texto):
    """
    Calcula el índice de coincidencia de las letras del texto
    múltiplos de la longitud de la clave
    """
    dic = {}
    alf = 'abcdefghijklmnopqrstuvwxyz'
    for k in range(1,50):
        clave = ''
        for j in range(k):
            clave += choice(alf)
        texto_cifrado = cifradoVigenere(texto,clave)
        dic[k] = indice_coincidencia(texto_cifrado)
    return dic   

    
def descifradoVigenere(texto,clave):
    """
    Devuelve el texto claro de Vigenère con la clave dada.
    """
    texto = arreglartexto(texto)
    clave = arreglartexto(clave)
    texto_claro = ''
    contador = 0
    longitud = len(clave)
    for caracter in texto:
        texto_claro += restar_caracteres(caracter,clave[contador%longitud])
        contador += 1
    return texto_claro
```
